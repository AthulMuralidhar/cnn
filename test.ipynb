# i want the loop to go from 00 01 10 11 then to 02 03 12 13 then to 20 21 30 31 then to 22 23 32 33

```python
import numpy as np
a = np.random.rand(4,4)
print(a)
ans1 = []
ans2 = []
ans3 = []
ans4 = []
for i in range(len(a)):
    for j in range(len(a)):
        if i <=1 and j <=1:
            ans1.append(a[i][j])
        if i<=1 and j>1:
            ans2.append(a[i][j])
        if i>1 & j<=1:
            ans3.append(a[i][j])
        if i>1 & j>1:
            ans4.append(a[i][j])
            
print(ans1)
print(ans2)
print(ans3)
print(ans4)
```
